## Importação de dataset via Pandas

```python
import pandas as pd

df = pd.read_csv("http://bit.ly/iris_ds",
                 names=['comp_sepala',
                        'larg_sepala',
                        'comp_petala',
                        'larg_petala',
                        'classe'])

df.head()
```

## Seleção de colunas com Pandas

```python
df_sepala = df[ ['comp_sepala', 'larg_sepala'] ]
```

## Seleção de linhas por índice com Pandas

```python
df_1_4 = df.iloc[ [1, 4] ] #Pega apenas as linhas 1 e 4
```

## Filtros de linhas por valores do DataFrame

Utilizando condições simples e múltiplas (AND e OR)

```python
df_m_5 = df [ df['comp_sepala'] > 5 ]
```

```python
df_and_5_5 = df[ (df['comp_sepala'] > 5) & (df['comp_petala'] > 5) ]
```

```python
df_or_5_5 = df[ (df['comp_sepala'] > 5) | (df['comp_petala'] > 5) ]
```

## Visualizando Informações

Gerando histograma de coluna

```python
df.hist();
```

Cruzando colunas com Seaborn

```python
import seaborn as sns

sns.scatterplot(x='comp_sepala',
                y='comp_petala',
                data=df,
                hue='classe');
```

```python
sns.pairplot(df, hue='classe');
```

Gerando Boxplot

```python
sns.boxplot(data=df, y='larg_petala');
```
