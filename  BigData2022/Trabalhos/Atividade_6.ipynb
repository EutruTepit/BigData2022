```python
!pip install apyori
from apyori import apriori
```

```python
compras = []

with open('store_data.csv') as c:
    for compra in c:
      items = compra.replace(' ', '').strip().split(',') # algums dados possui espaço logo após a ',' e outros n fazendo os dados msm representando a msm coisa o apriori 
                                                         # considerava como diff, retira a legibilidade humana e tem q fazer o tratamento antes de tentar comparar uma 
                                                         # nova compra com o modelo gerado
      compras.append(items)

compras
```

```python
suporte = 0.005
confianca = 0.10
lift=1.01
qtd=2

regras = apriori(compras, min_support=suporte, min_confidence=confianca, min_lift=lift, max_length=qtd)

regras = list(regras)
regras
```
