# Aula 04 - Gráficos do Seaborn

## Carregamento de dados

```python
import seaborn as sns
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv("https://bit.ly/iris_ds", names=['comp_sepala','larg_sepala','comp_petala','larg_petala','classe'])
df.head()
```

## Gráfico de Histograma e Densidade

Estes gráficos envolvem apenas uma variável por vez, que é a própria distribuição dos dados. Porém, pode-se utilizar para comparar mais variáveis.

displot

```python
sns.set()
sns.displot(data=df, x="larg_sepala");
```

Podemos plotar a densidade junto.

kde=True

```python
sns.displot(data=df, x="larg_sepala", hue="classe", kde=True);
```

Podemos ainda plotar identificando as classes

```python
sns.pairplot(df, hue="classe");
```

```python
sns.displot(data=df, x="larg_petala", hue="classe", kde=True);
```

Podemos ainda cruzar variáveis para verificar a densidade:

displot: x e y

```python
sns.displot(data=df, x="larg_petala", y="comp_petala");
```

Podemos ainda categorizar por classe, como nos outros gráficos

```python
sns.displot(data=df, x="larg_petala", y="comp_petala", hue="classe");
```

Podemos ainda plotar com base no KDE.

kind="kde"

```python
sns.displot(data=df, x="larg_petala", y="comp_petala", hue="classe", kind="kde");
```

## Gráfico de linhas

Gráfico de linhas (*lineplot*) é geralmente utilizado para visualização de dados ao longo do tempo (também chamado de série temporal - ou *time series*). A linha do tempo estaria disposta no eixo x, enquanto os dados devem estar no eixo y. 

Logo, sempre envolve duas variáveis: o tempo + outra variável de interesse.

No entanto, outras variáveis de interesse podem ser plotadas junto para comparação.

```python
sns.lineplot(data=df, y="comp_petala", x=np.arange(0, 150));
```

## Gráfico de barras

Gráfico de barras é utilizando quando precisamos avaliar o relacionamento entre uma variável categórica e uma variável numérica.

```python
sns.barplot(data=df, x="classe", y="comp_sepala");
```

Podemos ainda realizar a contagem de elementos por classe, por exemplo.

countplot

```python
sns.countplot(data=df, x="classe");
```

## Correlação

A correlação entre variáveis é importante para detecção de informações presentes em mais de uma variável simultaneamente.

```python
sns.heatmap(df.corr(), annot=True, fmt=".3f", vmin=-1);
```

## Boxplot

```python
sns.boxplot(data=df, x="larg_sepala");
sns.displot(data=df, x="larg_sepala");
```

```python
g = sns.jointplot(data=df, x="comp_petala", y="larg_sepala", hue="classe");
g.plot_marginals(sns.boxplot);
```

## LMPlot

LMPlot plota dados e um modelo regressivo ao mesmo tempo.

```python
sns.lmplot(data=df, x="comp_petala", y="comp_sepala");
```

```python
sns.lmplot(data=df, x="comp_petala", y="comp_sepala", hue="classe");
```
